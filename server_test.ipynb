```python
import socket
import random
import pickle


s = socket.socket()
host = socket.gethostname()
port = 7547
s.bind((host, port))

def storeData(obj, filename):
    pickleFile = open(filename, 'wb')
    pickle.dump(obj, pickleFile)
    pickleFile.close()

def loadData(filename):
    pickleFile = open(filename, 'rb')
    obj = pickle.load(pickleFile)
    pickleFile.close()
    return obj

response = loadData('response_request.dat')
request_list = loadData('request_url_list.dat')
```

```python
import socket
import random
import pickle
import requests
import time

from datetime import datetime

def storeData(obj, filename):
    pickleFile = open(filename, 'wb')
    pickle.dump(obj, pickleFile)
    pickleFile.close()

def loadData(filename):
    pickleFile = open(filename, 'rb')
    obj = pickle.load(pickleFile)
    pickleFile.close()
    return obj
```

```python
IoTIP = '192.168.3.39'
IoTPort = '80'


response_list = loadData('response_content.dat')
request_url_list = loadData('request_url_list.dat')
response_header = loadData('response_header.dat')
response_status = loadData('response_status.dat')

study_list = loadData('study_list.dat')
request_url_list.extend(study_list)

for i in study_list:
    time.sleep(1)
    request_url = 'http://'+ IoTIP + ':' + IoTPort + i
    try:
        r = requests.get(request_url, timeout = 3)
    except Timeout:
        print('The request timed out, so no response')
    else:
        a= b''
        for chunk in r.iter_content(chunk_size=128):
            a = a + chunk
        response_list[i] = a
        response_header[i] = r.headers
        response_status[i] = r.status_code
        print('request: ' + i+' done')
 
print('ALL Done')

storeData(response_list,"response_content.dat")
storeData(response_header,"response_header.dat")
storeData(response_status,"response_status.dat")
storeData(request_url_list,"request_url_list.dat")
```
